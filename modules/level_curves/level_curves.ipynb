The purpose of this module is plot visualize level curves of surfaces $z = f(x,y)$ in 3D space.

# $f(x,y) = x^2 + y^2$ 

```python
# x-y extent of domain
x_min = -1
x_max = 1
y_min = -1
y_max = 1

# extent of level set height
c_min = 0 
c_max = 1

# grid size
num_x = 100
num_y = 100
num_c = 11

def f(x, y):
  return x**2 + y**2
```

```python
make_figure(f, x_min, x_max, y_min, y_max, c_min, c_max, 
  num_x = num_x, num_y = num_y, num_c = num_c,
  plane_opacity = 0.9, surface_opacity = 0.5,
  level_curve_width = 3, level_curve_color = "black",
  intersection_width = 3, intersection_color = "green",
  filename = "bowl.html", save_figure = True)
```

```python
plot_contours(f, x_min, x_max, y_min, y_max, c_min, c_max, num_x=num_x, num_y = num_y, num_c = num_c)
```
